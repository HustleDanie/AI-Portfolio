```python
!pip install streamlit

import os
import streamlit as st
import numpy as np
import tensorflow as tf
from PIL import Image

# ---------------------------
# Helper: Train & Save the Model
# ---------------------------
def train_and_save_model():
    st.info("Training the MNIST model. This may take a few minutes...")
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

    # Normalize images to [0,1] and flatten them
    x_train = x_train.astype("float32") / 255.0
    x_test  = x_test.astype("float32") / 255.0
    x_train = x_train.reshape(-1, 28 * 28)
    x_test  = x_test.reshape(-1, 28 * 28)

    # Build a simple feed-forward network (MLP)
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(128, activation='relu', input_shape=(28*28,)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer="adam",
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])

    # Train the model
    history = model.fit(x_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=2)

    # Evaluate on test data
    loss, acc = model.evaluate(x_test, y_test, verbose=0)
    st.success(f"Model trained with test accuracy: {acc*100:.2f}%")

    # Save the trained model
    model.save("mnist_model.h5")
    st.info("Model saved as 'mnist_model.h5'")
    return model

# ---------------------------
# Helper: Load the Model (with caching)
# ---------------------------
@st.cache(allow_output_mutation=True)
def load_model():
    if not os.path.exists("mnist_model.h5"):
        model = train_and_save_model()
    else:
        model = tf.keras.models.load_model("mnist_model.h5")
        st.success("Pre-trained model loaded.")
    return model

# ---------------------------
# Helper: Image Preprocessing
# ---------------------------
def preprocess_image(image: Image.Image) -> np.ndarray:
    """
    Convert an uploaded image to a normalized 28x28 grayscale flattened array.
    """
    image = image.convert("L")            # Ensure image is grayscale
    image = image.resize((28, 28))          # Resize to 28x28 pixels
    img_array = np.array(image).astype("float32") / 255.0  # Normalize pixel values
    img_array = img_array.reshape(1, 28*28)  # Flatten to shape (1, 784)
    return img_array

# ---------------------------
# Streamlit App UI
# ---------------------------
st.title("MNIST Digit Classifier")
st.write("This app uses a feed-forward neural network to classify handwritten digits (0-9) from the MNIST dataset.")

# Load the model (trains if not already saved)
model = load_model()

# Option: Choose input method (upload image or random test image)
input_method = st.radio("Select input method:", ("Upload Image", "Random MNIST Test Image"))

if input_method == "Upload Image":
    uploaded_file = st.file_uploader("Upload an image file (PNG, JPG, JPEG) of a digit:", type=["png", "jpg", "jpeg"])
    if uploaded_file is not None:
        try:
            # Preprocess and display the image
            image = Image.open(uploaded_file)
            st.image(image, caption="Uploaded Image", use_column_width=True)
            img_array = preprocess_image(image)

            # Predict the digit
            prediction = model.predict(img_array)
            predicted_digit = np.argmax(prediction, axis=1)[0]
            st.write(f"**Predicted Digit: {predicted_digit}**")
        except Exception as e:
            st.error(f"Error processing the image: {e}")
    else:
        st.info("Please upload an image to get a prediction.")
else:
    if st.button("Show Random MNIST Test Image"):
        # Load MNIST test set directly
        (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
        x_test = x_test.astype("float32") / 255.0  # Normalize
        idx = np.random.randint(0, x_test.shape[0])
        random_img = x_test[idx]
        true_label = y_test[idx]

        # Display the random image with its true label
        st.image(random_img, caption=f"Random Test Image (True Label: {true_label})", width=150)

        # Flatten the image for prediction
        img_array = random_img.reshape(1, 28*28)
        prediction = model.predict(img_array)
        predicted_digit = np.argmax(prediction, axis=1)[0]
        st.write(f"**Predicted Digit: {predicted_digit}**")
```
