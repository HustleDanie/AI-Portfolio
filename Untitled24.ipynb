```python
# Write the app code to a file
%%writefile app.py
import os
import streamlit as st
import numpy as np
import tensorflow as tf
from PIL import Image

# ----------------------------
# MODEL TRAINING AND LOADING
# ----------------------------
def train_and_save_model():
    """
    Trains a feed-forward neural network on MNIST and saves the model.
    This function is called only if a pre-trained model is not found.
    """
    st.info("Training new MNIST model. This may take a few minutes...")
    # Load MNIST data
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
    # Normalize and flatten the images
    x_train = x_train.astype("float32") / 255.0
    x_test  = x_test.astype("float32") / 255.0
    x_train = x_train.reshape(-1, 28*28)
    x_test  = x_test.reshape(-1, 28*28)

    # Define a robust MLP model
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(256, activation="relu", input_shape=(28*28,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dense(10, activation="softmax")
    ])

    # Compile the model with Adam optimizer and appropriate loss for classification
    model.compile(optimizer="adam",
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])

    # Train the model; use 10% of the training data for validation
    history = model.fit(x_train, y_train,
                        epochs=10,
                        batch_size=128,
                        validation_split=0.1,
                        verbose=2)

    # Save the trained model to disk
    model.save("mnist_model.h5")
    st.success("Model training complete and saved as 'mnist_model.h5'.")
    return model

@st.cache(allow_output_mutation=True)
def load_model():
    """
    Loads a pre-trained MNIST model if it exists; otherwise, trains and saves a new one.
    Caching ensures that the model is loaded only once during the Streamlit session.
    """
    if not os.path.exists("mnist_model.h5"):
        model = train_and_save_model()
    else:
        model = tf.keras.models.load_model("mnist_model.h5")
        st.info("Loaded pre-trained model from disk.")
    return model

# ----------------------------
# IMAGE PREPROCESSING FUNCTION
# ----------------------------
def preprocess_uploaded_image(uploaded_file):
    """
    Preprocesses an uploaded image:
      - Converts to grayscale,
      - Resizes to 28x28,
      - Normalizes pixel values,
      - Flattens into a vector.
    """
    try:
        image = Image.open(uploaded_file).convert("L")
        image = image.resize((28, 28))
        img_array = np.array(image).astype("float32") / 255.0
        img_array = img_array.reshape(1, 28*28)
        return image, img_array
    except Exception as e:
        st.error(f"Error processing image: {e}")
        return None, None

# ----------------------------
# STREAMLIT FRONTEND
# ----------------------------
st.title("MNIST Digit Classifier")
st.write("This app uses a feed-forward neural network to classify handwritten digits (0-9).")

# Load the model (this call is cached for efficiency)
model = load_model()

# User input option: Upload an image or use a random MNIST test image.
option = st.radio("Choose your input method:", ("Upload Your Image", "Show Random MNIST Image"))

if option == "Upload Your Image":
    uploaded_file = st.file_uploader("Upload a 28x28 grayscale image (PNG/JPG)", type=["png", "jpg", "jpeg"])
    if uploaded_file is not None:
        # Preprocess the uploaded image
        image, img_array = preprocess_uploaded_image(uploaded_file)
        if image is not None:
            st.image(image, caption="Uploaded Image", width=150)
            # Predict using the pre-trained model
            prediction = model.predict(img_array)
            pred_digit = np.argmax(prediction, axis=1)[0]
            st.write(f"**Predicted Digit:** {pred_digit}")
    else:
        st.info("Please upload an image to see the prediction.")
else:
    if st.button("Show Random MNIST Test Image"):
        # Load MNIST test data
        (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
        x_test = x_test.astype("float32") / 255.0  # Normalize pixel values

        # Randomly select one test image
        idx = np.random.randint(0, x_test.shape[0])
        random_img = x_test[idx]
        true_label = y_test[idx]

        # Display the random image
        st.image(random_img, caption=f"Random Test Image (True Label: {true_label})", width=150)

        # Preprocess the random image: flatten and reshape
        img_array = random_img.reshape(1, 28*28)
        # Predict using the model
        prediction = model.predict(img_array)
        pred_digit = np.argmax(prediction, axis=1)[0]
        st.write(f"**Predicted Digit:** {pred_digit}")

```
